## Definition of Inheritance

___

## class Child: public Parent
## class GrandChild: public Child

___

Child and grand child has access to parent protected and public members

```c++
class Parent{
private:
    int a;
protected:
    int b;
public:
    int c;

public:
    
    Parent(){
       cout<<"Non Param Parent"<<endl;     
        a=1;
        b=2;
        c=3;
    }
    
    
    Parent(int n){
        cout<<"Param Parent:"<<n<<endl;     
    }
};

class Child:public Parent{
private:
    int a_1;
protected:
    int b; //Since its protected above
    int b_1;
public:
    
    int c_1;

    
public:
    
    Child(){
       cout<<"Non Param Child"<<endl;     
        a_1=4;
        b_1=5;
        c_1=6;
    }
    
    
    Child(int n){
        cout<<"Param Child:"<<n<<endl;     
    }
    
    Child(int a, int b):Parent(a){
       cout<<"Param Child (Special):"<<a<<" "<<b<<endl;     
        
    }
    
};


class GrandChild:public Child{
private:
    int a_2;
protected:
    int b_2;
public:
    int c_2;

    
public:
    
    GrandChild(){
       cout<<"Non Param GrandChild"<<endl;     
    }
    
    
    GrandChild(int n){
        cout<<"Param GrandChild:"<<n<<endl;     
    }
   
};
```
