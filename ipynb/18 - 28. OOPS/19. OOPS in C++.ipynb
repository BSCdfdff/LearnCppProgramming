+ Defining a object (of Class) in STACK
+ Using pointer

````c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
````

````c++
class Rectangle{
public:    
    int length;
    int breadth;
    
    int area(){
        return length * breadth;
    }
    
    int perimeter(){
        return 2 *( length + breadth);
    }
    
}
````

## Creating pointer to Class object (in STACK)


````c++
Rectangle r1;
Rectangle *p;
````

````c++
p = &r1;
````

````c++
r1.length=10;
r1.breadth=5;
cout<<r1.length<<endl;
cout<<r1.breadth<<endl;
````

## Stucture pointer deference

Same as \*p where we need to get to value of data variable

We use -> to get to value, but this is also in place of (.) for operator

````c++
p->length=10;
p->breadth=5;
cout<<p->length<<endl;
cout<<p->breadth<<endl;
````

````c++
cout<<p->area()<<endl;
````

## <center>Pointer in Stack</center>


## How does it work

We first set the data variable object of Custom Data type of **Rectangle**, namely **r1**.

```
Rectangle r1

```

Then we define a pointer of type Rectangle, **\*p**

```
Rectangle *p

```


Then we set **p** equal to the address of the object data variable

```
p = &r1

```

So previously in in order to get the data (values) using the pointer, we had to use the (*) dereferencing symbol.

Now we use the **stucture pointer deference** or **arrow** symbol

```
->

```
Ad it is used **AFTER** p.

p->length
p->breadth

Which coincidently is in the place of (.) dot operator, which can cause confusion.

With the Pointer Dereferencing (->) you can access (of Class):

1. Data Memeber
2. Functions
