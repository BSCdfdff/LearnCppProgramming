+ Defining a Class

## What is a class and how does it work

**Refering to the DEMO code below***

+ User Defined data type
+ We can then defined an instance of if by defining an object of this type
+ These variables (objects), are also created in STACK


````c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
````

````c++
class Rectangle{
public:    
    int length;
    int breadth;
    
    int area(){
        return length * breadth;
    }
    
    int perimeter(){
        return 2 *( length + breadth);
    }
    
}
````

````c++
Rectangle r1, r2;
````

## How much memory space (bytes) will this object occupy?

We need to look at the data members (the data variables defined in class) that is used in class:

```
int length;
int breadth;
int area(){..}
int perimeter(){}
```

int takes 4 bytes x 2 (lenght and breath) = 8 bytes.

Note functions will not occupy any memeory space.

So size of this class (Rectangle) = 8 bytes.

```
r1 = 8 bytes
r2 = 8 bytes
```

Note the class, which is the blueprint, does NOT occupy any memory, only the instance, i.e the object occupies memory

````c++
r1.length=10;
r1.breadth=5;
````

## How to access member of class

We use the dot (.) operator

But by default all member are private.

So we need to add **public:**, and every thing after this declaration becomes accessible from outside the class

```
public:    
    int length;
    int breadth;
    ....
    ....
```

(This is how the learning will occur: only when things are needed we add it, so I will remember it better next time. By adding public from start would not have made sense. This wil be the learning approach taken)



````c++
cout<<r1.area()<<endl;
cout<<r1.perimeter()<<endl;
````

### Remember functions will NOT occupy any memory

````c++
r2.length=15;
r2.breadth=10;
````

````c++
cout<<r2.area()<<endl;
cout<<r2.perimeter()<<endl;
````
