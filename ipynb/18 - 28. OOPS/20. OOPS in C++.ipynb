+ Defining a object (of Class) in HEAP
+ Using pointer

````c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
````

````c++
class Rectangle{
public:    
    int length;
    int breadth;
    
    int area(){
        return length * breadth;
    }
    
    int perimeter(){
        return 2 *( length + breadth);
    }
    
}
````

## Creating pointer in HEAP


````c++
Rectangle *p;
````

````c++
p = new Rectangle
````

````c++
p->length=10;
p->breadth=5;
cout<<p->length<<endl;
cout<<p->breadth<<endl;
````

````c++
cout<<p->area()<<endl;
````

````c++
Rectangle *q = new Rectangle()
````

````c++
q->length=15;
q->breadth=10;
````

````c++
cout<<q->area()<<endl;
````

## <center>Pointer in HEAP</center>


## How does it work

Then we DECLARE a pointer of type Rectangle, **\*p**

```
Rectangle *p

```

No need to define a data varianle first            

Then we INITIALIZE te pointer using **new** keyword, to signify the pointer needs to be in HEAP

```
p = new Rectangle
```

We could have also written it in ONE line

```
Rectangle *q = new Rectangle()

```

The rest is the same as STACK, logic

Onlt here we need to deallocate memory, when we done with the variables


## Lets recap...

Declaring the variable (object) in STACK:

```
Rectangle r

```

Declaring the pointer for data variable (address) in STACK:

```
Rectangle r
Rectangle *p;
p = &r
```



Declaring the pointer in HEAP:

```
Rectangle *q = new Rectangle()
```
