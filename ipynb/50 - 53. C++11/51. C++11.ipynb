## Smart pointers

We do know when creating pointers in HEAP, we need to de-allocate it afterwards

So c++ will deallocate memory when function not in use:

There are three smart pointer available:

1. unique_ptr
2. shared_ptr
3. weak_ptr


## Unique Ptr




It can only point on **ONE** object at time (we can remove, then add another)

Remember these are template pointers, so it needs to be followed by type.

We will be creating the object in HEAP (new keyword)

Once ptr defined, we **CANNOT** do

```
unique_ptr<Rectangle> ptr2 = ptr;

```

But we can remove ptr pointing rectangle and have ptr2 point to it, that is **ALLOWED**

```
unique_ptr<Rectangle> ptr2;
ptr2=move(ptr);

```

If we try this, we'll get **error***, and this **ptr** now is **null** (managed by c++)

```
c<<ptr->area()<<e;
```

````c++
#include <iostream>
#define c cout
#define e endl
using namespace std;
````

````c++
class Rectangle{
    int length;
    int breadth;
public: 
    Rectangle(int l, int b){
        length = l;
        breadth = b;
    }
    int area(){
        return length * breadth;
    }    
}
````

````c++
unique_ptr<Rectangle> ptr(new Rectangle(10,5));
````

````c++
c<<ptr->area()<<e;
````

````c++
unique_ptr<Rectangle> ptr2;
````

````c++
ptr2=move(ptr);
````

````c++
c<<ptr2->area()<<e;
````

## Shared Ptr




Now we can share so the following will work

```
shared_ptr<Rectangle> ptr_s2;
ptr_s2 = ptr_s;

```

````c++
shared_ptr<Rectangle> ptr_s(new Rectangle(10,5));
````

````c++
c<<ptr_s->area()<<e;
````

````c++
shared_ptr<Rectangle> ptr_s2;
````

````c++
ptr_s2 = ptr_s;
````

````c++
c<<ptr_s->area()<<e;
````

````c++
c<<ptr_s2->area()<<e;
````

````c++
c<<ptr_s.use_count()<<e;
````

````c++
c<<ptr_s2.use_count()<<e;
````

## Weak Ptr




Works similar to sharing, but dont maintain a ref_counter.

This is to avoid deadlocks, and its a weaker implementation
