##  Exception Handling

Here we talking about run time errors, i.e Errors which are caused by the user

Not Logical (tested using debugger) or Syntax (tested using compiler)

Pupose of Exceptions: **To give proper Error messages to the user**

c++ does **NOT** have a mechanism to throw an exeption, we have to not only throw it **ourselves**, but also **CHECK** for the error as well, to know when to throw the error.

```
if (b==0)
   throw 1;

```

But the reason why it works this way (***where you need to check and throw the error yourself***), e.g. 

I have a **division function**, that needs to perform division, but lets say you pass a zero, to divide by

Then its the purpose of the division function to **throw error**, NOT Return Error, to the calling function:

```

int division (int x, int y){
    if (y==0)
        throw 1;
    }
    return x/y;
```

And since we throwing an **int**, we need to define to define the catch as **int**:

```
catch(int e){
    ...
}

```
So the link is **throw** and **catch**, and the **1** we use to throw it will be the **e** value


### Variations of Exeption Handling
___

#### Multiple throw value and catch blocks

Previously we throw int value. Can we throw float, double, char, string, ect? Yes, can throw anything.

We can even throw our own class:

```
class MyException{}

if (b==)
    throw MyException();

```
And you can or dont have to inherit from exception, e.g:

```
class MyException: public exception

```

Also the function that you declaring, you can define in the signature that the function is throwing it in its signature:

```
int division (int x, int y) throw (MyException) {

    if (b==)
        throw MyException;
    }

}
```


or thowing an int then 

```
int division (int x, int y) throw (int) {

    if (b==0)
        throw 1;
    }

}
```

But the signature in the function definition is not mandatory, its just some information to calling program to show that this function wil throw the type of errors in throw signature.

But then we need to check for it?

But it we have the following, that means this function does not throw exception, as the throw is empty:

```
int division (int x, int y) throw () {

    return x/y

```


#### Catch "ALL" 

```
catch(...){
    cout<<"Some Error Occurred"<<"endl;
}

```

But ```catch(...)``` must be the **last** catch block

We can also have nested try/catch inside each other


Where we inherit Execeptions the flow has to be from Child to Parent, i.e Parent has to be defined last:

```
class Exception1{}

class Exception2: public Exception1{}

try{
...
}
catch (Exception2 e){}
catch (Exception1 e){}
```


````c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
````

````c++
int a=10; 
int b=0; 
int c;

try{
    if (b==0)
        throw 1;
    c = a/b;
    cout<<c<<endl;
}
catch(int e){
    cout<<"Division by Zero (int)"<<"--Error code:"<<e<<endl;
}
catch(float e){
    cout<<"Division by Zero (float)"<<"--Error code:"<<e<<endl;
}
catch(double e){
    cout<<"Division by Zero (double)"<<"--Error code:"<<e<<endl;
}
catch(char e){
    cout<<"Division by Zero (char)"<<"--Error code:"<<e<<endl;
}
catch(string e){
    cout<<"Division by Zero (string)"<<"--Error code:"<<e<<endl;
}
// catch(...){
//     cout<<"Some Error Occurred"<<"endl;
// }
cout<<"Bye"<<endl;
````
