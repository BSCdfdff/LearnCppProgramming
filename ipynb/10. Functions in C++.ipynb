## How does it work?


Remember a pointer reference is just an alias or nickname of the variable we referencing.

The main point that we can see from the above drawing, and that **NO ADDITIONAL MEMORY** is created by swap function...

Also, how does swap function then change the values of the parameters of main function...lets explore...

As normal, declare and initialize x=10, y=20, and pass it to swap:

```
swap(10,20);

```

But now we update the signature of swap function:

```
void swap (int &a, int &b)

```

What does the above mean? Whichever function made a call to swap, ie. main, it will use the same addresses of parameters, i.e. x and y, main is using.

So, its will not create extra memory addresses for a, b, but will use the addresses of x, y

But we know RULE: One functions cannot access another functions parameter (i.e passed arguments), unless through pointer. But swap does not have a pointers to main (its parameters). So how is this possible?

The answer is how the compiler generates machine code: Previously when swap is called, swap gets its own activation record. i.e. in code section of memory, swap is not part of main code.

Now, although the source code has a swap and main function(s), the machine code generated by the compiler, makes copy (inserts) of swap into main, .i.e. it becomes part of main. Swap will also be referred to an INLINE function

Since it now part of main, it no longer a separate function, so main/swap is one function, as swap can access the address of x,y

No other language has this ability (as per author)

#### Advantage:

+ No additional memory is created

But it comes with a cautionary: Do no put complex logic, e.g. loops into CALLED (swap) function, as the compiler may not correctly get it right. (???)

So main take away: You actually working with the calling parameter (x,y), i.e. a, b is just alias/nickname for x,y

## Pass by Reference

````c++
#include <iostream>
#include <climits>
#include <math.h>
using namespace std;
````

````c++
void swap(int &a, int &b){
    int temp;
    temp=a;
    a=b;
    b=temp;
}
````

````c++
int x = 10, y = 20;
cout<<x<< " " <<y<<endl;
swap(x,y);
cout<<x<< " " <<y<<endl;
````
